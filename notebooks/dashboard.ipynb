```python
import ipywidgets as widgets
from IPython.display import display, clear_output

def check_uid_wid(wid_value, uid_value):
    flask_url = f"https://data-delinquents-db-9edf072e759e.herokuapp.com/database-delinquents/user/getID"
    data = {"uid": uid_value, 'wid': wid_value}
    response = requests.post(flask_url, json=data)
    return response

# Header Styling
header_text = widgets.HTML("<h1 style='text-align: center; color: white; background-color: #333; width: 100%; margin: 0; padding: 15px 0;'>Inventory App</h1>")


dropdown_options = [
    '-------',
    "Warehouse’s year profit",
    "Top 5 racks with quantity under the 25% capacity threshold",
    "Bottom 3 part’s type/material in the warehouse",
    "Top 5 most expensive racks in the warehouse",
    "Top 3 supplier that supplied to the warehouse",
    "Top 3 days with the smallest incoming transactions’ cost",
    "Top 3 users that receives the most exchanges",
    "See all transactions in a warehouse (Sorted by latest to oldest)",
    "See all parts prices.",
    "See all parts supplied by suppliers.",
    "See all parts in warehouses."
]

global_options = [
    '-------',
    "Top 10 warehouses with the most racks",
    "Top 5 warehouses with the most incoming transactions",
    "Top 5 warehouses that delivers the most exchanges",
    "Top 3 users that made the most transactions",
    "Top 3 warehouses with the least outgoing transactions",
    "Top 3 warehouses’ cities with the most transactions"
]


dropdown = widgets.Dropdown(options=dropdown_options, value='-------', description='Select:')

dropdown = widgets.Dropdown(options=dropdown_options, description='Dropdown:', style={'width': '150px', 'background-color': '#f7f7f7', 'border-radius': '5px', 'padding': '8px'}, disabled = True)

dropdown2 = widgets.Dropdown(options=global_options, value='-------', description='Select:')

dropdown2 = widgets.Dropdown(options=global_options, description='Global Select:', style={'width': '150px', 'background-color': '#f7f7f7', 'border-radius': '5px', 'padding': '8px'})

# Text Input Styling
input_1 = widgets.Text(placeholder='Warehouse ID', layout=widgets.Layout(width='200px', margin='5px'))
input_2 = widgets.Text(placeholder='User ID', layout=widgets.Layout(width='200px', margin='5px'))
input_3 = widgets.Text(placeholder='Supplier ID', layout=widgets.Layout(width='200px', margin='5px'))
button = widgets.Button(description="Log In")

# Input box Styling
input_box = widgets.VBox([input_1, input_2, button, dropdown, input_3, dropdown2], layout=widgets.Layout(border='1px solid black', border_radius='5px', background_color='#f7f7f7', margin='5px', padding='8px', width='18%'))

# Output widget to display values
output = widgets.Output()


logged_in = False  # Flag to track login/logout state

def on_button_click(b):
    global logged_in
    with output:
        clear_output()
        if not logged_in:
            title_box_html = f'''
                <div style="background-color: #FFE087; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">Loading ...</h2>
                </div>
            '''
            display(HTML(title_box_html))
            value = check_uid_wid(input_1.value, input_2.value)
            if value == 'False':
                clear_output()
                title_box_html = '''<div style="background-color: #D22B2B; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">User Doesn't Work on Warehouse</h2>
                </div>
                '''
                display(HTML(title_box_html))
                time.sleep(2)
                clear_output()
            else:
                clear_output()
                title_box_html = '''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">Logged In, Welcome!</h2>
                </div>
                '''
                display(HTML(title_box_html))
                dropdown.disabled = False
                input_1.disabled = True
                input_2.disabled = True
                button.description = "Log Out"  
                logged_in = True  
                time.sleep(2)
                clear_output()
        else:
            dropdown.disabled = True
            input_1.disabled = False
            input_2.disabled = False
            button.description = "Log In!"  
            logged_in = False  



# Attach the function to the button's click event
button.on_click(on_button_click)


# Displaying Header, Dropdown, and Input Box
display(header_text)
display(input_box)
display(output)


# Function to run when the dropdown value changes
def on_dropdown_change(change):
    with output:
        clear_output()
        new_value = change['new']


        if new_value == "Warehouse’s year profit":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            warehouse_year_profit(input_1.value, input_2.value)
        elif new_value == "Top 5 racks with quantity under the 25% capacity threshold":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_racks_under_capacity(input_1.value, input_2.value)
        elif new_value == "Bottom 3 part’s type/material in the warehouse":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            bottom_parts_type_material(input_1.value, input_2.value)
        elif new_value == "Top 5 most expensive racks in the warehouse":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_expensive_racks(input_1.value, input_2.value)
        elif new_value == "Top 3 supplier that supplied to the warehouse":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_supplier_supplied(input_1.value, input_2.value)
        elif new_value == "Top 3 days with the smallest incoming transactions’ cost":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_days_smallest_cost(input_1.value, input_2.value)
        elif new_value == "Top 3 users that receives the most exchanges":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_users_most_exchanges(input_1.value, input_2.value)
        elif new_value == "See all transactions in a warehouse (Sorted by latest to oldest)":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            see_all_transactions(input_1.value)
        elif new_value == "See all parts prices.":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            see_all_parts_prices()
        elif new_value == "See all parts supplied by suppliers.":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            if input_3.value == '':
                title_box_html = '''<div style="background-color: black; padding: 10px; text-align: center; margin: 20px auto; width: 45%;">
                    <h2 style="margin: 0; color: white;">Supplier ID Missing</h2>
                </div>
                '''
                display(HTML(title_box_html))
                time.sleep(1)
                clear_output()
                return 0
            see_all_parts_supplied(input_3.value)
        elif new_value == "See all parts in warehouses.":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            see_all_parts_warehouses(input_1.value)

def on_dropdown2_change(change):
    with output:
        clear_output()
        new_value = change['new']

        if new_value == "Top 10 warehouses with the most racks":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_10_most_racks()
        elif new_value == "Top 5 warehouses with the most incoming transactions":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_5_most_incoming_transactions()
        elif new_value == "Top 5 warehouses that delivers the most exchanges":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_5_most_delivers_exchanges()
        elif new_value == "Top 3 users that made the most transactions":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_3_users_most_transactions()
        elif new_value == "Top 3 warehouses with the least outgoing transactions":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_3_warehouses_least_outgoing_transactions()
        elif new_value == "Top 3 warehouses’ cities with the most transactions":
            title_box_html = f'''<div style="background-color: #005b96; padding: 10px; text-align: center; margin: 20px auto; width: 40%;">
                    <h2 style="margin: 0; color: white;">{new_value}</h2>
                </div>
                '''
            display(HTML(title_box_html))
            top_3_cities_most_transactions()




dropdown.observe(on_dropdown_change, names='value')

dropdown2.observe(on_dropdown2_change, names='value')
```
